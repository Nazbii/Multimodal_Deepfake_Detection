### Video Processing Setup
Тhe dataset is downloaded and the necessary libraries are installed (or at least attempted),  can start processing the video files. A common first step for face and lip-sync analysis is to extract individual frames from the videos.

```python
import os
from moviepy.editor import VideoFileClip

# DATASET_BASE_PATH will be defined later after dataset download

# Helper function to extract frames from a video
def extract_frames(video_path, output_folder, frame_rate=1):
    """
    Extracts frames from a video file and saves them to an output folder.

    Args:
        video_path (str): Path to the input video file.
        output_folder (str): Path to the folder where frames will be saved.
        frame_rate (int): Number of frames per second to extract.
    """
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    try:
        clip = VideoFileClip(video_path)
        # Save frames at the specified frame_rate
        for i, frame in enumerate(clip.iter_frames(fps=frame_rate, dtype='uint8')):
            frame_filename = os.path.join(output_folder, f"frame_{i:06d}.jpg")
            # Using pillow directly for saving since moviepy's default writer can be slow
            import PIL.Image
            PIL.Image.fromarray(frame).save(frame_filename)
        clip.close()
        print(f"Extracted frames from {video_path} to {output_folder}")
    except Exception as e:
        print(f"Error extracting frames from {video_path}: {e}")
```

The `extract_frames` function will take a video file, an output directory, and an optional frame rate. It will save JPEG images of the video frames into the specified folder. This is crucial for subsequent face detection and analysis.

### Face Detection and Cropping

After extracting frames, the next step for face-based deepfake analysis is to detect and crop faces from each frame. This helps to normalize the input for subsequent models and focuses on the most relevant part of the image.

```python
import torch
from facenet_pytorch import MTCNN, InceptionResnetV1
import cv2
import numpy as np

# Initializing MTCNN for face detection
#  We are using the 'keep_all=True' to detect multiple faces, though for deepfakes we often expect one main face.
mtcnn = MTCNN(image_size=160, margin=0, min_face_size=20, thresholds=[0.6, 0.7, 0.7], factor=0.709, post_process=True, device='cuda' if torch.cuda.is_available() else 'cpu')

def detect_and_crop_faces(input_frames_folder, output_faces_folder):
    """
    Detects faces in images within input_frames_folder and saves cropped faces to output_faces_folder.
    """
    if not os.path.exists(output_faces_folder):
        os.makedirs(output_faces_folder)

    print(f"Detecting and cropping faces from {input_frames_folder}...")
    frame_files = sorted([f for f in os.listdir(input_frames_folder) if f.endswith('.jpg')])

    for i, frame_file in enumerate(frame_files):
        frame_path = os.path.join(input_frames_folder, frame_file)
        img = cv2.imread(frame_path)
        if img is None:
            print(f"Warning: Could not read image {frame_path}. Skipping.")
            continue

        # Convert BGR to RGB (OpenCV reads in BGR)
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # Detect faces
        boxes, _ = mtcnn.detect(img_rgb)

        if boxes is not None:
            for j, box in enumerate(boxes):

                # Convert box coordinates to integers
                box = [int(b) for b in box]
                x1, y1, x2, y2 = box

                # Ensure coordinates are within image bounds
                h, w, _ = img.shape
                x1, y1, x2, y2 = max(0, x1), max(0, y1), min(w, x2), min(h, y2)

                # Crop face
                cropped_face = img[y1:y2, x1:x2]

                if cropped_face.size == 0:
                    print(f"Warning: Cropped face is empty for {frame_file} box {j}. Skipping.")
                    continue

                # Save cropped face
                face_filename = f"face_{i:06d}_{j:02d}.jpg"
                face_path = os.path.join(output_faces_folder, face_filename)
                cv2.imwrite(face_path, cropped_face)
        else:
            print(f"No face detected in {frame_file}")
    print(f"Finished processing faces for {input_frames_folder}. Output saved to {output_faces_folder}")
```

The `detect_and_crop_faces` function takes an input directory of frames, detects faces using MTCNN, and saves the cropped face images to an output directory. Let's apply this to our sample frames.

### Mouth Region of Interest (ROI) Extraction

```python
# Install mediapipe if not already installed
!pip install mediapipe

# IMPORTANT: After running this cell, please restart your Colab runtime (Runtime -> Restart runtime)
# This is necessary for the updated mediapipe version to be recognized by all loaded libraries.
```

```python
import mediapipe as mp
import cv2
import numpy as np
import os

# Initialize MediaPipe Face Mesh
mp_face_mesh = mp.solutions.face_mesh

def detect_mouth_roi(image_path, output_folder, image_size=96):
    """
    Detects the mouth region in a face image using MediaPipe Face Mesh
    and saves the cropped mouth ROI.

    Args:
        image_path (str): Path to the input face image.
        output_folder (str): Path to the folder where cropped mouth ROIs will be saved.
        image_size (int): Desired size for the square mouth ROI image.
    """
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    image = cv2.imread(image_path)
    if image is None:
        print(f"Warning: Could not read image {image_path}. Skipping.")
        return

    # Convert the BGR image to RGB.
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    with mp_face_mesh.FaceMesh(
        static_image_mode=True,
        max_num_faces=1,
        refine_landmarks=True,
        min_detection_confidence=0.5) as face_mesh:

        results = face_mesh.process(image_rgb)

        if not results.multi_face_landmarks:
            # print(f"No face landmarks detected in {image_path}. Skipping.")
            return

        for face_landmarks in results.multi_face_landmarks:
            # Get landmarks for the mouth region (outer lips)
            mouth_landmarks_indices = [
                61, 146, 91, 181, 84, 17, 314, 405, 321, 375, # Upper outer lip
                78, 191, 80, 81, 82, 13, 312, 311, 310, 415, # Lower outer lip
            ]

            x_coords = []
            y_coords = []
            img_h, img_w, _ = image.shape

            for idx in mouth_landmarks_indices:
                if idx < len(face_landmarks.landmark):
                    landmark = face_landmarks.landmark[idx]
                    x_coords.append(int(landmark.x * img_w))
                    y_coords.append(int(landmark.y * img_h))

            if not x_coords or not y_coords:
                # print(f"Could not get valid mouth landmarks for {image_path}. Skipping.")
                return

            # Calculate bounding box around mouth landmarks with padding
            padding = 10
            min_x = max(0, min(x_coords) - padding)
            max_x = min(img_w, max(x_coords) + padding)
            min_y = max(0, min(y_coords) - padding)
            max_y = min(img_h, max(y_coords) + padding)

            # Ensure the ROI is square for consistent input
            width = max_x - min_x
            height = max_y - min_y

            if width > height:
                diff = width - height
                min_y -= diff // 2
                max_y += diff - (diff // 2)
            elif height > width:
                diff = height - width
                min_x -= diff // 2
                max_x += diff - (diff // 2)

            # Re-check bounds after making it square
            min_x = max(0, min_x)
            max_x = min(img_w, max_x)
            min_y = max(0, min_y)
            max_y = min(img_h, max(y_coords))

            # Crop the mouth ROI
            mouth_roi = image[min_y:max_y, min_x:max_x]

            if mouth_roi.size == 0:
                # print(f"Warning: Cropped mouth ROI is empty for {image_path}. Skipping.")
                return

            # Resize to a consistent size (e.g., 96x96 for SyncNet)
            mouth_roi_resized = cv2.resize(mouth_roi, (image_size, image_size))

            # Save cropped mouth ROI
            base_name = os.path.basename(image_path)
            mouth_roi_filename = f"mouth_roi_{base_name}"
            mouth_roi_path = os.path.join(output_folder, mouth_roi_filename)
            cv2.imwrite(mouth_roi_path, mouth_roi_resized)
            # print(f"Saved mouth ROI for {image_path} to {mouth_roi_path}")

print("Mouth ROI detection function defined.")
```

The `detect_mouth_roi` function uses MediaPipe's Face Mesh to identify key facial landmarks. It then calculates a bounding box around the outer lip landmarks, adds a small padding, and crops this region. The cropped mouth ROI is resized to a consistent square size (defaulting to 96x96 pixels, which is common for models like SyncNet) and saved as a new image file. This process isolates the most relevant visual information for lip-sync analysis.

```python
# Assuming `output_faces_folder` from the previous step is still valid
# Need to re-run the face detection cell if `output_faces_folder` is not defined.

if 'output_faces_folder' in locals() and os.path.exists(output_faces_folder):
    output_mouth_roi_folder = os.path.join('/content/extracted_mouth_rois_new', os.path.basename(output_faces_folder))
    if not os.path.exists(output_mouth_roi_folder):
        os.makedirs(output_mouth_roi_folder)

    print(f"Detecting and cropping mouth ROIs from {output_faces_folder}...")
    face_image_files = sorted([f for f in os.listdir(output_faces_folder) if f.startswith('face_') and f.endswith('.jpg')])

    for face_file in face_image_files:
        face_path = os.path.join(output_faces_folder, face_file)
        detect_mouth_roi(face_path, output_mouth_roi_folder)

    print(f"Finished processing mouth ROIs for {output_faces_folder}. Output saved to {output_mouth_roi_folder}")

    # Verify extraction by listing contents of the output folder
    print(f"\nContents of {output_mouth_roi_folder}:")
    !ls {output_mouth_roi_folder}
else:
    print("Error: `output_faces_folder` is not defined or does not exist. Please run the face detection cell first.")
```

### Audio Extraction

For multimodal deepfake detection, especially for lip-syncing analysis, we need to extract the audio track from the video files. This separated audio can then be processed independently.

The `extract_audio` function extracts the audio track from a given video file and saves it as a WAV file. This function uses `ffmpeg` for robust audio processing, ensuring compatibility and quality for subsequent audio analysis steps.

```python
import os
import subprocess # To run shell commands

def extract_audio(video_path, output_audio_path):
    """
    Extracts audio from a video file and saves it as an audio file using ffmpeg.

    Args:
        video_path (str): Path to the input video file.
        output_audio_path (str): Path to save the extracted audio file (e.g., .wav, .mp3).
    """
    try:
        # Checking if the video file actually exists
        if not os.path.exists(video_path):
            print(f"Error: Video file not found at {video_path}")
            return

        # Using ffmpeg directly for more robust audio extraction
        # -y overwrites existing file without asking
        # -vn disables video recording
        # -acodec pcm_s16le sets the audio codec to uncompressed PCM
        # -ar 44100 sets the audio sampling rate
        # -ac 1 sets the number of audio channels to mono
        command = [
            'ffmpeg', '-y',
            '-i', video_path,
            '-vn',
            '-acodec', 'pcm_s16le',
            '-ar', '44100',
            '-ac', '1',
            output_audio_path
        ]
        subprocess.run(command, check=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)

        print(f"Extracted audio from {video_path} to {output_audio_path}")
    except subprocess.CalledProcessError as e:
        print(f"Error extracting audio from {video_path} using ffmpeg: {e}")
        print(f"FFmpeg stdout: {e.stdout.decode()}")
        print(f"FFmpeg stderr: {e.stderr.decode()}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
```

The `extract_mfccs` function processes an audio file to compute Mel-frequency cepstral coefficients (MFCCs). MFCCs are a powerful feature representation for audio, especially speech, widely used in tasks like speech recognition and audio deepfake detection because they effectively capture the spectral characteristics of sound in a compact form.

```python
import librosa
import numpy as np

def extract_mfccs(audio_path, n_mfcc=13, hop_length=512, sr=16000):
    """
    Extracts MFCC features from an audio file.

    Args:
        audio_path (str): Path to the input audio file.
        n_mfcc (int): Number of MFCCs to return.
        hop_length (int): The hop-length for MFCC calculation.
        sr (int): Target sampling rate for audio.

    Returns:
        np.ndarray: MFCC features, or None if an error occurs.
    """
    try:
        audio, sample_rate = librosa.load(audio_path, sr=sr)
        mfccs = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=n_mfcc, hop_length=hop_length)
        print(f"Extracted MFCCs from {audio_path}. Shape: {mfccs.shape}")
        return mfccs
    except Exception as e:
        print(f"Error extracting MFCCs from {audio_path}: {e}")
        return None
```

The `extract_mfccs` function will load an audio file, resample it to a target sampling rate, and then compute the Mel-frequency cepstral coefficients (MFCCs). MFCCs are widely used in speech recognition and audio analysis to represent the short-term power spectrum of a sound. We will apply this to our extracted sample audio.

```python
# Assuming `output_audio_path` from the previous step is still valid
# Need to re-run the audio extraction cell if it's not defined.

if 'output_audio_path' in locals() and os.path.exists(output_audio_path):
    print(f"Processing audio file: {output_audio_path}")
    sample_mfccs = extract_mfccs(output_audio_path)

    if sample_mfccs is not None:
        print(f"Sample MFCCs successfully extracted with shape: {sample_mfccs.shape}")
    else:
        print("Failed to extract MFCCs from the sample audio.")
else:
    print("Error: `output_audio_path` is not defined or does not exist. Please run the audio extraction cell first.")
```

### Exploring Diffusion Models

Diffusion models are a class of generative models that have shown impressive results in generating high-quality images from text prompts or other inputs. They work by iteratively denoising a random noise image until it becomes a coherent output. We can explore how to integrate these into the project, potentially for generating synthetic deepfake data or for understanding the generative process itself.

First, we need to install the `diffusers` library, which provides pre-trained diffusion models and tools to use them easily. We'll also need `transformers` for tokenization and model loading.

Now, let's load a pre-trained diffusion model, such as Stable Diffusion, and use it to generate an image based on a simple text prompt. We will use the `pipeline` API from `diffusers` for simplicity.

```python
from diffusers import DiffusionPipeline
import torch
import matplotlib.pyplot as plt

# Load a pre-trained Stable Diffusion pipeline
# Using 'stabilityai/stable-diffusion-xl-base-1.0' as an example
# This might take a while to download the first time.
# Ensure you have enough GPU memory if running on a GPU.

pipeline = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16)
pipeline.to("cuda" if torch.cuda.is_available() else "cpu")

# Define a text prompt
prompt = "a photo of a high-tech robot dog in a futuristic city"

# Generate an image
image = pipeline(prompt).images[0]

# Display the generated image
plt.imshow(image)
plt.axis('off')
plt.title("Generated Image from Diffusion Model")
plt.show()

print("Image generated successfully!")
```
